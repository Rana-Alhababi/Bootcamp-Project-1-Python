```python
# Personal Finance Tracker

import matplotlib.pyplot as plt

# List to store financial transactions
transactions = []

# Function to add a transaction
def add_transaction():
    category = input("Enter the category: ")
    amount = float(input("Enter the amount: "))
    date = input("Enter the date (YYYY-MM-DD): ")

    transaction = {"category": category, "amount": amount, "date": date}
    transactions.append(transaction)
    print("Transaction added successfully!")

# Function to calculate total expenses
def calculate_total_expenses():
    total_expenses = sum(transaction["amount"] for transaction in transactions if transaction["amount"] < 0)
    return total_expenses

# Function to calculate total income
def calculate_total_income():
    total_income = sum(transaction["amount"] for transaction in transactions if transaction["amount"] > 0)
    return total_income

# Function to generate a summary report
def generate_summary_report():
    total_expenses = calculate_total_expenses()
    total_income = calculate_total_income()
    net_balance = total_income + total_expenses

    print("Summary Report:")
    print("Total Expenses:", total_expenses)
    print("Total Income:", total_income)
    print("Net Balance:", net_balance)

# Function to sort transactions by amount using a Lambda function
def sort_transactions_by_amount():
    sorted_transactions = sorted(transactions, key=lambda transaction: transaction["amount"])
    return sorted_transactions

# Function to categorize transactions
def categorize_transactions():
    categories = set(transaction["category"] for transaction in transactions)
    print("Available Categories:")
    for category in categories:
        print(category)

# Function to generate monthly expense report
def generate_monthly_report():
    monthly_expenses = {}
    for transaction in transactions:
        date_parts = transaction["date"].split("-")
        month = date_parts[1]
        if month in monthly_expenses:
            monthly_expenses[month] += transaction["amount"]
        else:
            monthly_expenses[month] = transaction["amount"]

    # Sort the monthly expenses by month
    sorted_expenses = sorted(monthly_expenses.items(), key=lambda x: int(x[0]))

    # Generate the report
    print("Monthly Expense Report:")
    for month, expense in sorted_expenses:
        print("Month:", month)
        print("Total Expenses:", expense)
        print("-" * 20)

# Function to visualize spending patterns
def visualize_spending_patterns():
    categories = []
    amounts = []

    for transaction in transactions:
        categories.append(transaction["category"])
        amounts.append(abs(transaction["amount"]))

    # Create a pie chart
    plt.pie(amounts, labels=categories, autopct="%1.1f%%")
    plt.axis("equal")
    plt.title("Spending Patterns")
    plt.show()

# Main program loop
while True:
    print("Personal Finance Tracker")
    print("1. Add Transaction")
    print("2. Calculate Total Expenses")
    print("3. Calculate Total Income")
    print("4. Generate Summary Report")
    print("5. Sort Transactions by Amount")
    print("6. Categorize Transactions")
    print("7. Generate Monthly Expense Report")
    print("8. Visualize Spending Patterns")
    print("9. Exit")

    choice = input("Enter your choice (1-9): ")

    if choice == "1":
        add_transaction()
    elif choice == "2":
        total_expenses = calculate_total_expenses()
        print("Total Expenses:", total_expenses)
    elif choice == "3":
        total_income = calculate_total_income()
        print("Total Income:", total_income)
    elif choice == "4":
        generate_summary_report()
    elif choice == "5":
        sorted_transactions = sort_transactions_by_amount()
        print("Sorted Transactions by Amount:")
        for transaction in sorted_transactions:
            print(transaction)
    elif choice == "6":
        categorize_transactions()
    elif choice == "7":
        generate_monthly_report()
    elif choice == "8":
        visualize_spending_patterns()
    elif choice == "9":
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please try again.")
```
